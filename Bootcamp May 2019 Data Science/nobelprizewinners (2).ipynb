## 7. Repeat laureates
<p>For most scientists/writers/activists a Nobel Prize would be the crowning achievement of a long career. But for some people, one is just not enough, and few have gotten it more than once. Who are these lucky few? (Having won no Nobel Prize myself, I'll assume it's just about luck.)</p>

```python
# Selecting the laureates that have received 2 or more prizes.
# ... YOUR CODE FOR TASK 7 ...full_name

multiple_winners = nobel.groupby('full_name').full_name.count()
print(multiple_winners[multiple_winners.between(2,multiple_winners.max())])
print('There are '+str(multiple_winners[multiple_winners.between(2,multiple_winners.max())].count())+' laureates that have received 2 or more prizes.')
```

-  We create multiple winners dataframe grouping the entries by full_name getting them to count the number of occurences in the nobel dataframe. WE can then tell it to only get the entries between 2 values one being 2 as the minimum as that means ahas won it more than once.
-  The other parameter is multiple_winners.max() to get the highest number of wins so its scalable regardless of number of wins
-  We can the count that series and see the number of entries

## 10. Oldest and youngest winners
<p>In the peace category there is a winner around 2010 that seems exceptionally young. This begs the questions, who are the oldest and youngest people ever to have won a Nobel Prize?</p>

```python
ageWinners = nobel[nobel.age.isnull() != True] 

ageWinners= ageWinners.sort_values(by=['age'],ascending=False)

print(ageWinners['full_name'].iloc[0]+ ' is the oldest as the age '+str(ageWinners['age'].iloc[0]))
print(ageWinners['full_name'].iloc[-1]+ ' is the oldest as the age '+str(ageWinners['age'].iloc[-1]))
```

-  We  create a ageWinners dataframe and store results that arent NaN hence the nobel.age.isnull != True  
-  We then sort the values by age. and assign that to the ageWinners dataframe changing the order of ages from highest to lowest using ascending = False
-  We the get the 'full_name' and 'age' using the iloc index of -1 for youngest and 0 as oldest ,as the -1 is last entry.
